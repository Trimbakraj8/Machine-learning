```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif, SelectPercentile
from sklearn.decomposition import PCA

```

```python
# Load the iris dataset
data = load_iris()
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, random_state=42)
```

```python
# Convert target arrays to integer arrays
y_train = y_train.astype(int)
y_test = y_test.astype(int)
```

```python
# Use SelectKBest to select the top 2 features
selector = SelectKBest(score_func=f_classif, k=2)
X_train_kbest = selector.fit_transform(X_train, y_train)
X_test_kbest = selector.transform(X_test)
```

```python
# Use SelectPercentile to select the top 50% features
selector = SelectPercentile(percentile=50)
X_train_percentile = selector.fit_transform(X_train, y_train)
X_test_percentile = selector.transform(X_test)
```

```python
# Print the selected features
selected_features = np.array(data.feature_names)[selector.get_support()]
print("SelectKBest selected the following features: ")
print(selected_features)
```

```python
# Use SelectPercentile to select the top 50% features
selector = SelectPercentile(percentile=50)
X_train_percentile = selector.fit_transform(X_train, y_train)
X_test_percentile = selector.transform(X_test)
```

```python
# Print the selected features
selected_features = np.array(data.feature_names)[selector.get_support()]
print("SelectPercentile selected the following features: ")
print(selected_features)
```

```python
# Use PCA to reduce the dimensionality to 2
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# Display results
print("Original Shape: ", data.data.shape)
print("SelectKBest")
print("SelectKBest train Shape: ", X_train_kbest.shape)
print("SelectKBest test Shape: ", X_test_kbest.shape)
print("SelectPercentile")
print("SelectPercentile train Shape: ", X_train_percentile.shape)
print("SelectPercentile test Shape:  ", X_test_percentile.shape)
print("PCA")
print("PCA train Shape: ", X_train_pca.shape)
print("PCA test Shape:  ",X_test_pca.shape)
```
